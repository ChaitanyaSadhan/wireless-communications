```python
#importing of required modules or libraries.
import numpy as np
import numpy.random as nr
import matplotlib.pyplot as plt
```

```python

nSamples = 1000000 #setting number of samples as 10 lacs

#generating fading channel coefficients.
h = (nr.normal(0, 1, nSamples)+1j*nr.normal(0, 1, nSamples) ) /np.sqrt(2);
#and they are divided with square root of 2 to normalize the variance

a = np.abs(h); #magnitudes of the complex noise samples.
phi = np.angle(h); #phases of the comples noise samples.
x= np.arange(0,4,0.01);
theta = np.arange(-np.pi,np.pi,0.01);
nbins = 1000;
```

```python
#PDF of Amplitude of Channel
plt.figure(1)
plt.hist(a,bins=nbins, density=True) #plotting the histogram
#density is kept true to normalise the histogram (to get pdf in otherwords.)
plt.plot(x,2*x*np.exp(-x**2),'r') #plotting rayleigh pdf

#labeling and titling and adding legends to the plot
plt.suptitle('PDF of Amplitude')#title of the plot
plt.xlabel('x') #x label
plt.ylabel('$f_A$(a)') #y label
#text between the dollar signs is interpreted as mathematical equation
plt.legend(["PDF Plot","Histogram"],loc ="upper right") #legend on the top right
```

#**Observations**
###The pdf of magnitude of the noise samples is rayleigh pdf.

```python
#PDF of Phase of Channel
plt.figure(2)

#plotting of the pdf of the phase
plt.hist(phi,bins=nbins, density=True)
plt.plot(theta, 1/2/np.pi*np. ones (np. shape (theta)) , 'r') #plotting uniform distribution

#labeling and adding titles and legends to the plot
plt.suptitle('PDF of Phase')#title of the plot
plt.xlabel('$\phi$') #x label
plt.ylabel('$f_\Phi(\phi)$')#y label
#the text between the dollar signs are interpreted as mathematical equations
plt.legend(["PDF Plot","Histogram"],loc ="upper right")
```

#**Observations**
###The pdf of phase of the noise samples is uniform distribution.
